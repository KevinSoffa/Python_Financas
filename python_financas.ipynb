# Pegando cotações da IBOVESP de 2022

```python
import pandas_datareader.data as web
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


cotacao_ibov = web.DataReader('^BVSP', data_source='yahoo', start='2022-01-01', end='2022-11-10')
display(cotacao_ibov)
```

## Analise | Gráfico

```python
cotacao_ibov['Adj Close'].plot(figsize=(15, 5))
```

## Retorno do IBOV

```python
retorno_ibov = cotacao_ibov['Adj Close'][-1] / cotacao_ibov['Adj Close'][0] - 1
print('Retorno IBOV: {:.2%}'.format(retorno_ibov))
```

## Analise de Média Móvel

```python
cotacao_ibov['Adj Close'].plot(figsize=(15, 5), label='IBOV')
cotacao_ibov['Adj Close'].rolling(21).mean().plot(label='Média Movel 21')
plt.legend()
plt.show()
```
